```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

obj = {"class" : ["A","B","A","C","E","D","A","F"]}
df= pd.DataFrame(obj)



encoder = LabelEncoder()
encoder.fit(df["class"]) #encoder.fit(df["Class"])
labels = encoder.transform(df["class"])

df["to_수치형데이터"]=labels

df
```

```python
obj2 = {
    "과일" : ["사과", "포도" ,"배" , "감","딸기","수박","사과", "배"]
}

df2 = pd.DataFrame(obj2)

encoder = LabelEncoder()
encoder.fit(df2)
labels2 = encoder.transform(df2)

df2["to_수치데이터"] = labels2

df2
```
