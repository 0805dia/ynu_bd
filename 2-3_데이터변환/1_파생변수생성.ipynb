```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(iris.data,columns = iris.feature_names)
data.insert(0,column = "ID", value = range(150))
data.head()
```

```python
# 구간화
#cut()

cut_data=pd.cut(data["sepal length (cm)"], bins =5 )
data["cut_data"]= cut_data
data.head(10)
```

```python
cut_data_2 = pd.cut(data["sepal length (cm)"], bins =[4,5,6,7,8 ])
data["컷데이터"] = cut_data_2
data.head()
```

```python
#qcut

pd.qcut(data["sepal length (cm)"], q = 4)


print(data["sepal length (cm)"].quantile(.25))
print(data["sepal length (cm)"].quantile(.5))
print(data["sepal length (cm)"].quantile(.75))

print(data["petal length (cm)"].max())
print(data["petal length (cm)"].min())
```
